```python
class Process:
    def __init__(self, bt):
        self.at = 0  # Arrival Time
        self.bt = bt  # CPU Burst time
        self.ct = 0   # Completion Time
        self.wt = 0   # Waiting Time
        self.tat = 0  # Turnaround Time
        self.rt = 0   # Response Time
        self.start_time = 0  # Start Time


def srtf(ps):
    n = len(ps)
    bt_remaining = [p.bt for p in ps]  # Initialize bt_remaining with burst times from ps
    is_completed = [False] * n
    is_first_process = True
    current_time = 0
    completed = 0
    sum_tat = 0
    sum_wt = 0
    sum_rt = 0
    total_idle_time = 0
    length_cycle = 0
    prev = 0

    while completed != n:
        min_index = -1
        minimum = float("inf")

        for i in range(n):
            if ps[i].at <= current_time and not is_completed[i]:
                if bt_remaining[i] < minimum:
                    minimum = bt_remaining[i]
                    min_index = i
                if bt_remaining[i] == minimum:
                    if ps[i].at < ps[min_index].at:
                        minimum = bt_remaining[i]
                        min_index = i

        if min_index == -1:
            current_time += 1
        else:
            if bt_remaining[min_index] == ps[min_index].bt:
                ps[min_index].start_time = current_time
                total_idle_time += ps[min_index].start_time - prev
                is_first_process = False

            bt_remaining[min_index] -= 1
            current_time += 1
            prev = current_time

            if bt_remaining[min_index] == 0:
                ps[min_index].ct = current_time
                ps[min_index].tat = ps[min_index].ct - ps[min_index].at
                ps[min_index].wt = ps[min_index].tat - ps[min_index].bt
                ps[min_index].rt = ps[min_index].start_time - ps[min_index].at

                sum_tat += ps[min_index].tat
                sum_wt += ps[min_index].wt
                sum_rt += ps[min_index].rt
                completed += 1
                is_completed[min_index] = True

    max_completion_time = float("-inf")
    min_arrival_time = float("inf")

    for i in range(n):
        max_completion_time = max(max_completion_time, ps[i].ct)
        min_arrival_time = min(min_arrival_time, ps[i].at)

    length_cycle = max_completion_time - min_arrival_time

    cpu_utilization = (length_cycle - total_idle_time) / length_cycle

    return sum_tat / n ,sum_wt / n ,sum_rt / n

def Norm(results, distributions):
    normalized_results = {}
    for distribution in results:
        avg_turnaround_times = [result[0] for result in results[distribution]]
        avg_waiting_times = [result[1] for result in results[distribution]]
        avg_response_times = [result[2] for result in results[distribution]]

        normalized_turnaround = [(t - min(avg_turnaround_times)) / (max(avg_turnaround_times) - min(avg_turnaround_times)) for t in avg_turnaround_times]
        normalized_waiting = [(w - min(avg_waiting_times)) / (max(avg_waiting_times) - min(avg_waiting_times)) for w in avg_waiting_times]
        normalized_response = [(r - min(avg_response_times)) / (max(avg_response_times) - min(avg_response_times)) for r in avg_response_times]

        normalized_results[distribution] = (normalized_turnaround, normalized_waiting, normalized_response)

    # Plotting the normalized results as range box plots
    plt.figure(figsize=(12, 6))
    plt.title('Normalized Average Times - SRTF Scheduling')
    plt.ylabel('Normalized Time')
    plt.xlabel('Distribution')
    labels = ['Turnaround Time', 'Waiting Time', 'Response Time']

    for i, label in enumerate(labels):
        data = [normalized_results[distribution][i] for distribution in distributions]
        plt.boxplot(data, positions=np.arange(len(distributions)) + 1, widths=0.6, showfliers=True)

    plt.xticks(np.arange(len(distributions)) + 1, distributions)
    plt.legend(labels)
    plt.show()

def foo():
    distributions = ['Normal', 'Exponential', 'Geometric', 'Negative Binomial', 'Poisson', 'Uniform']
    results = {}

    for distribution in distributions:
        results[distribution] = []

        for _ in range(100):
            np.random.seed(42 + _)
            burst_times = []

            if distribution == 'Normal':
                burst_times = np.random.normal(250, 100, size=100)
            elif distribution == 'Exponential':
                burst_times = np.random.exponential(scale=250, size=100)
            elif distribution == 'Geometric':
                burst_times = np.random.geometric(p=0.2, size=100)
            elif distribution == 'Negative Binomial':
                burst_times = np.random.negative_binomial(n=5, p=0.3, size=100)
            elif distribution == 'Poisson':
                burst_times = np.random.poisson(lam=5, size=100)
            elif distribution == 'Uniform':
                burst_times = np.random.uniform(low=1, high=500, size=100)

            burst_times = np.clip(burst_times, 1, 500)
            proc = [Process(int(bt)) for bt in burst_times]
            aT, aW, aR = srtf(proc)
            results[distribution].append((aT, aW, aR))

    Norm(results, distributions)

foo()

```

```python
import numpy as np
import matplotlib.pyplot as plt

class Process:
    def __init__(self, bt):
        self.at = 0  # Arrival Time
        self.bt = bt  # CPU Burst time
        self.ct = 0   # Completion Time
        self.wt = 0   # Waiting Time
        self.tat = 0  # Turnaround Time
        self.rt = 0   # Response Time
        self.start_time = 0  # Start Time

def fcfs(ps):
    sum_tat = 0
    sum_wt = 0
    sum_rt = 0
    total_idle_time = 0
    length_cycle = 0
    cpu_utilization = 0
    n = len(ps)

    for i in range(n):
        ps[i].start_time = ps[i].at if i == 0 else max(ps[i].at, ps[i - 1].ct)
        ps[i].ct = ps[i].start_time + ps[i].bt
        ps[i].tat = ps[i].ct - ps[i].at
        ps[i].wt = ps[i].tat - ps[i].bt
        ps[i].rt = ps[i].wt

        sum_tat += ps[i].tat
        sum_wt += ps[i].wt
        sum_rt += ps[i].rt
        total_idle_time += 0 if i == 0 else (ps[i].start_time - ps[i - 1].ct)

    length_cycle = ps[n - 1].ct - ps[0].start_time
    cpu_utilization = (length_cycle - total_idle_time) / length_cycle

    return sum_tat / n, sum_wt / n, sum_rt / n

def Norm(results, distributions):
    normalized_results = {}
    for distribution in results:
        avg_turnaround_times = [result[0] for result in results[distribution]]
        avg_waiting_times = [result[1] for result in results[distribution]]
        avg_response_times = [result[2] for result in results[distribution]]

        normalized_turnaround = [(t - min(avg_turnaround_times)) / (max(avg_turnaround_times) - min(avg_turnaround_times)) for t in avg_turnaround_times]
        normalized_waiting = [(w - min(avg_waiting_times)) / (max(avg_waiting_times) - min(avg_waiting_times)) for w in avg_waiting_times]
        normalized_response = [(r - min(avg_response_times)) / (max(avg_response_times) - min(avg_response_times)) for r in avg_response_times]

        normalized_results[distribution] = (normalized_turnaround, normalized_waiting, normalized_response)

    # Plotting the normalized results as range box plots
    plt.figure(figsize=(12, 6))
    plt.title('Normalized Average Times - FCFS Scheduling')
    plt.ylabel('Normalized Time')
    plt.xlabel('Distribution')
    labels = ['Turnaround Time', 'Waiting Time', 'Response Time']

    for i, label in enumerate(labels):
        data = [normalized_results[distribution][i] for distribution in distributions]
        plt.boxplot(data, positions=np.arange(len(distributions)) + 1, widths=0.6, showfliers=True)

    plt.xticks(np.arange(len(distributions)) + 1, distributions)
    plt.legend(labels)
    plt.show()

def foo():
    distributions = ['Normal', 'Exponential', 'Geometric', 'Negative Binomial', 'Poisson', 'Uniform']
    results = {}

    for distribution in distributions:
        results[distribution] = []

        for _ in range(100):
            np.random.seed(42 + _)
            burst_times = []

            if distribution == 'Normal':
                burst_times = np.random.normal(250, 100, size=100)
            elif distribution == 'Exponential':
                burst_times = np.random.exponential(scale=250, size=100)
            elif distribution == 'Geometric':
                burst_times = np.random.geometric(p=0.2, size=100)
            elif distribution == 'Negative Binomial':
                burst_times = np.random.negative_binomial(n=5, p=0.3, size=100)
            elif distribution == 'Poisson':
                burst_times = np.random.poisson(lam=5, size=100)
            elif distribution == 'Uniform':
                burst_times = np.random.uniform(low=1, high=500, size=100)

            burst_times = np.clip(burst_times, 1, 500)
            proc = [Process(int(bt)) for bt in burst_times]
            aT, aW, aR = fcfs(proc)
            results[distribution].append((aT, aW, aR))

    Norm(results, distributions)

foo()
```

```python
import math
import statistics as stat
class Process:
    def __init__(self, bt):

        self.at = 0  # Arrival Time
        self.bt = bt  # CPU Burst time
        self.ct = 0   # Completion Time
        self.wt = 0   # Waiting Time
        self.tat = 0  # Turnaround Time
        self.rt = 0   # Response Time
        self.start_time = 0  # Start Time
        self.bt_remaining = bt

def rr(ps):
    n = len(ps)
    q = []
    visited = [False] * 100
    current_time = 0
    completed = 0
    tq = 0
    total_idle_time = 0
    sum_tat = 0
    sum_wt = 0
    sum_rt = 0


    for _ in ps:
        tq += _.bt
    tq = math.ceil(tq / n)


    q.append(0)
    visited[0] = True

    while completed != n:
        index = q.pop(0)

        if ps[index].bt_remaining == ps[index].bt:
            ps[index].start_time = max(current_time, ps[index].at)
            total_idle_time += ps[index].start_time - current_time
            current_time = ps[index].start_time

        if ps[index].bt_remaining - tq > 0:
            ps[index].bt_remaining -= tq
            current_time += tq
        else:
            current_time += ps[index].bt_remaining
            ps[index].bt_remaining = 0
            completed += 1

            ps[index].ct = current_time
            ps[index].tat = ps[index].ct - ps[index].at
            ps[index].wt = ps[index].tat - ps[index].bt
            ps[index].rt = ps[index].start_time - ps[index].at

            sum_tat += ps[index].tat
            sum_wt += ps[index].wt
            sum_rt += ps[index].rt

        for i in range(1, n):
            if ps[i].bt_remaining > 0 and ps[i].at <= current_time and not visited[i]:
                q.append(i)
                visited[i] = True

        if ps[index].bt_remaining > 0:
            q.append(index)

        if not q:
            for i in range(1, n):
                if ps[i].bt_remaining > 0:
                    q.append(i)
                    visited[i] = True
                    break

    max_completion_time = float("-inf")

    for i in range(n):
        max_completion_time = max(max_completion_time, ps[i].ct)

    length_cycle = max_completion_time - ps[0].at

    cpu_utilization = (length_cycle - total_idle_time) / length_cycle

    return (sum_tat / n), (sum_wt / n), (sum_rt / n)

def Norm(results, distributions):
    normalized_results = {}
    for distribution in results:
        avg_turnaround_times = [result[0] for result in results[distribution]]
        avg_waiting_times = [result[1] for result in results[distribution]]
        avg_response_times = [result[2] for result in results[distribution]]

        normalized_turnaround = [(t - min(avg_turnaround_times)) / (max(avg_turnaround_times) - min(avg_turnaround_times)) for t in avg_turnaround_times]
        normalized_waiting = [(w - min(avg_waiting_times)) / (max(avg_waiting_times) - min(avg_waiting_times)) for w in avg_waiting_times]
        normalized_response = [(r - min(avg_response_times)) / (max(avg_response_times) - min(avg_response_times)) for r in avg_response_times]

        normalized_results[distribution] = (normalized_turnaround, normalized_waiting, normalized_response)

    # Plotting the normalized results as range box plots
    plt.figure(figsize=(12, 6))
    plt.title('Normalized Average Times - RoundRobin Scheduling')
    plt.ylabel('Normalized Time')
    plt.xlabel('Distribution')
    labels = ['Turnaround Time', 'Waiting Time', 'Response Time']

    for i, label in enumerate(labels):
        data = [normalized_results[distribution][i] for distribution in distributions]
        plt.boxplot(data, positions=np.arange(len(distributions)) + 1, widths=0.6, showfliers=True)

    plt.xticks(np.arange(len(distributions)) + 1, distributions)
    plt.legend(labels)
    plt.show()

def foo():
    distributions = ['Normal', 'Exponential', 'Geometric', 'Negative Binomial', 'Poisson', 'Uniform']
    results = {}

    for distribution in distributions:
        results[distribution] = []

        for _ in range(100):
            np.random.seed(42 + _)
            burst_times = []

            if distribution == 'Normal':
                burst_times = np.random.normal(250, 100, size=100)
            elif distribution == 'Exponential':
                burst_times = np.random.exponential(scale=250, size=100)
            elif distribution == 'Geometric':
                burst_times = np.random.geometric(p=0.2, size=100)
            elif distribution == 'Negative Binomial':
                burst_times = np.random.negative_binomial(n=5, p=0.3, size=100)
            elif distribution == 'Poisson':
                burst_times = np.random.poisson(lam=5, size=100)
            elif distribution == 'Uniform':
                burst_times = np.random.uniform(low=1, high=500, size=100)

            burst_times = np.clip(burst_times, 1, 500)
            proc = [Process(int(bt)) for bt in burst_times]
            aT, aW, aR = rr(proc)
            results[distribution].append((aT, aW, aR))

    Norm(results, distributions)

foo()
```

```python
from math import ceil
from statistics import mean, median

# Set DEBUG to False to only see the analysis
# Set DEBUG to True to see steps in the run
DEBUG = False
def debug_print(text):
    if DEBUG:
        print(text)


class Clock:
    """
        Clock class mocks a wall clock.  Used to track start/end times easier than trying to imply by wait etc times.
    """
    def __init__(self):
        self._tick = 0
    def increment_tick(self, value):
        self._tick += value
    def get_tick(self):
        return self._tick


class Process:
    def __init__(self, label, burst_time):
        self.label = label
        self.burst_time = burst_time
        self.remaining_time = burst_time
        self.waiting_time = 0 # set to 0 because a 1 waiting_time is not by definition waiting_time
        self.preemption_count = 0  # set to 0 because a 1 preemption_count is not by definition preemption_count
        self.start_time = 0
        self.end_time = 0
        self.fairness_ratio = 0

    def run_for_quantum(self, clock, time_quantum):
        """
            run_for_quantum does all the changes need to the internal state of a process
                waiting_time,
                start_time
                end_time
                premption_count
            and to the clock handed in (increments by used time)
            returns the time used in the run so that managing process can update remaining time_quantum.
        """
        if self.burst_time == self.remaining_time:
            self.start_time = clock.get_tick()
        used_time = min(self.remaining_time, time_quantum)
        clock.increment_tick(used_time)
        self.remaining_time -= used_time
        self.preemption_count += 1
        if self.remaining_time == 0:
            self.end_time = clock.get_tick()
        debug_print(f"ran process {self.label} for {used_time}")
        return used_time

    def __str__(self):
        return f"[{self.label:15} burst_time:{self.burst_time:3} remaining_time:{self.remaining_time:3}" + \
            f" waiting_time:{self.waiting_time:3} start_time:{self.start_time:3} end_time:{self.end_time:3} fairness_ratio:{self.fairness_ratio:4.4}]"



def debug_print_batch(batch):
    debug_print("---- BATCH ----")
    for process in batch:
        debug_print(f"  {process}")
    debug_print("---------------")



def run_quantum(clock, batch, time_quantum):
    # run processes until run out of quantum
    # update wait time each time.
    debug_print(f"Running Quantum @ {clock.get_tick()} with {time_quantum=} with batch")
    debug_print_batch(batch)
    for process in batch:
        if time_quantum <= 0:
            break
        used_time = process.run_for_quantum(clock, time_quantum)
        # update waiting time for all other processes that waited for this process to run
        # including process that ran earlier in this time quantum
        for waiting_process in batch:
            if waiting_process != process and waiting_process.remaining_time > 0:
                waiting_process.waiting_time += used_time
                debug_print(f"updated wait time for process={waiting_process.label:15} by {used_time:3} to {waiting_process.waiting_time:3}")
            else:
                debug_print(f"skipped wait time for process={waiting_process.label:15} since we just ran it or it is done")
        # update the time quantum with the amoutnrun
        time_quantum -= used_time

    for process in batch[:]:
        if process.remaining_time == 0:
            debug_print(f"removing finished process {process}")
            batch.remove(process)
    debug_print("Finished Quantum")


# these are the core aspects of the algorithm
# the rest (minus the sort) is just a classic premptive scheduling framework
# and this could all be pulled out to a class defining the parameter of the algo so it could
# support any arbitrary scheduling algo like RR, etc...
# the 1 + is to avoid all processes starting with ratio of 0
# so at the begining the ratio is equivelent to 1/process.burst_time
def calculate_fairness_ratio(process):
    return (1 + process.burst_time - process.remaining_time + process.waiting_time) / (process.burst_time * (1 + process.preemption_count))


def calc_time_quantum(batch):
    average_remaining_time = mean(process.remaining_time for process in batch)
    median_remaining_time = median(process.remaining_time for process in batch)
    return ceil((average_remaining_time + median_remaining_time) / 2)


def algorithm(batch):
    wall_clock = Clock()
    batch = batch[:] #copy list since we modify it.
    while len(batch):
        for process in batch:
            process.fairness_ratio = calculate_fairness_ratio(process)

        batch.sort(key=lambda process: process.fairness_ratio, reverse=True)
        time_quantum = calc_time_quantum(batch)

        run_quantum(wall_clock, batch, time_quantum)


def analyze(batch):
    debug_print("-----Analyze----")
    debug_print_batch(batch)

    average_turnaround_time = mean([process.end_time for process in batch])
    average_waiting_time = mean([process.waiting_time for process in batch])
    average_response_time = mean([process.start_time for process in batch])

    print("Average turnaround time:", average_turnaround_time)
    print("Average waiting time:", average_waiting_time)
    print("Average response time:", average_response_time)


if __name__ == "__main__":
    distributions = ['Normal', 'Exponential', 'Geometric', 'Negative Binomial', 'Poisson', 'Uniform']
    results = {}

    for distribution in distributions:

        results[distribution] = []

        for _ in range(100):
            np.random.seed(42+_)
            batch = []
            burst_times = []

            if distribution == 'Normal':
                burst_times = np.random.normal(250, 100, size=100)
            elif distribution == 'Exponential':
                burst_times = np.random.exponential(scale=250, size=100)
            elif distribution == 'Geometric':
                burst_times = np.random.geometric(p=0.2, size=100)
            elif distribution == 'Negative Binomial':
                burst_times = np.random.negative_binomial(n=5, p=0.3, size=100)
            elif distribution == 'Poisson':
                burst_times = np.random.poisson(lam=5, size=100)
            elif distribution == 'Uniform':
                burst_times = np.random.uniform(low=1, high=500, size=100)

            burst_times = np.clip(burst_times, 1, 500)
            for j, burst_time in enumerate(burst_times):
                burst_time = int(burst_time)
                batch.append(Process(f"Process {j}", burst_time))

            algorithm(batch)
            results[distribution].append((
                mean([process.end_time for process in batch]),
                mean([process.waiting_time for process in batch]),
                mean([process.start_time for process in batch])
            ))

    # Normalizing the data
    normalized_results = {}
    for distribution in results:
        avg_turnaround_times = [result[0] for result in results[distribution]]
        avg_waiting_times = [result[1] for result in results[distribution]]
        avg_response_times = [result[2] for result in results[distribution]]

        normalized_turnaround = [(t - min(avg_turnaround_times)) / (max(avg_turnaround_times) - min(avg_turnaround_times)) for t in avg_turnaround_times]
        normalized_waiting = [(w - min(avg_waiting_times)) / (max(avg_waiting_times) - min(avg_waiting_times)) for w in avg_waiting_times]
        normalized_response = [(r - min(avg_response_times)) / (max(avg_response_times) - min(avg_response_times)) for r in avg_response_times]

        normalized_results[distribution] = (normalized_turnaround, normalized_waiting, normalized_response)

    # Plotting the normalized results as range box plots
    plt.figure(figsize=(12, 6))
    plt.title('Normalized Average Times - FairBatch Scheduling')
    plt.ylabel('Normalized Time')
    plt.xlabel('Distribution')
    labels = ['Turnaround Time', 'Waiting Time', 'Response Time']

    for i, label in enumerate(labels):
        data = [normalized_results[distribution][i] for distribution in distributions]
        plt.boxplot(data, positions=np.arange(len(distributions)) + 1, widths=0.6, showfliers=True)

    plt.xticks(np.arange(len(distributions)) + 1, distributions)
    plt.legend(labels)
    plt.show()
```
